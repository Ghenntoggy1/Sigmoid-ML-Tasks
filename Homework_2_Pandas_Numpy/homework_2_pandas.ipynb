Next, I will check the missing values in the dataset using isnull() method from pandas, that will show a boolean for each cell (True if Null, False otherwise), and combine it with

Next, I will pass through each of the columns and get rid of Non-Numerical Columns or Columns that has no effect on the Survived column, that may be logically explained without
evidence, such as: PassengerId, because it is a unique identifier for each row in the dataset, Name, Ticket. About Cabin, I found that Cabin Code is the placement on the Ship, more
exactly, the Deck where the Cabin was placed, therefore most probably there is a correlation between the placement of the cabin and the survival rate of the passenger, but I might
just drop it as well, since I suppose that the PClass is related with the Cabin placement a well. I will try to just take the first letters of the Cabin, for example
Passenger with cabin C85, will be assigned to the group with C Cabin, thus I will maybe find a correlation with the actual status if Passenger survived or not.

```python
dataset = dataset.drop(["PassengerId", "Name", "Ticket"], axis=1)
dataset = dataset.dropna(subset=["Cabin"])
dataset
```

Next, I will sum up the Siblings and Parents/Children together, to make a new column, that will hold the family size altogether.

```python
dataset["FamilySize"] = dataset["SibSp"] + dataset["Parch"] + 1
dataset.drop(["SibSp", "Parch"], axis=1, inplace=True)
dataset
```

```python
copy_dataset = dataset.copy()
```

Next, I will do the above said, I will take only the Code of the Cabin, thus making it more Categorical Data than it is right now.

```python
dataset["Cabin"] = dataset["Cabin"].apply(lambda cabin_name: cabin_name[0])
dataset
```

Next, I will find all unique values in the dataset for each column. First, I will drop the remaining Null values from the dataset, in Age and Embarked.

```python
dataset.dropna(subset=["Age", "Embarked"], inplace=True)
dataset
```

```python
original_dataset = dataset.copy()
```

Then, I will apply the unique() function in order to obtain all the uniques values in each column. This will enable me to correctly encode / map the values that are string to
integers, such as 0 and 1 and so on.

```python
unique_values = dataset.apply(pd.Series.unique)
unique_values = unique_values.apply(lambda x : sorted(x))
print(unique_values)
```

As it may be seen, there are several columns that may be mapped to integers. I will be using map() method in order to map each Category to a number.

```python
dataset["Sex"] = dataset["Sex"].map({"female": 0, "male": 1})
dataset["Deck"] = dataset["Cabin"].map({"A": 7, "B": 6, "C": 5, "D": 4, "E": 3, "F": 2, "G": 1, "T": 0})
dataset.drop(columns=["Cabin"], axis=1, inplace=True)
dataset["Embarked"] = dataset["Embarked"].map({"C": 0, "Q": 1, "S": 2})
dataset
```
